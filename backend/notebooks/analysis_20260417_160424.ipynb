```python

        import pandas as pd
        import os

        # Robust path searching for both automation and manual debugging
        filename = "POP_SalesTransactionHistory.csv"
        candidates = [
            os.path.join("public", "data", filename),           # Path from project root
            os.path.join("..", "..", "public", "data", filename) # Path from backend/notebooks/
        ]

        file_path = None
        for path in candidates:
            if os.path.exists(path):
                file_path = path
                break

        if file_path:
            df_sales = pd.read_csv(file_path)
            print(f"Successfully loaded {file_path} into df_sales")
            print(f"Columns in df_sales: {list(df_sales.columns)}")
            print(f"\nDataFrame Info for df_sales:")
            df_sales.info()
            print(df_sales.head())
        else:
            print(f"Error: File {filename} not found in expected locations: {candidates}")
            print(f"Current working directory: {os.getcwd()}")
        
```

```python
import pandas as pd

# Calculate total sales amount
df_sales['Sales_Amount'] = df_sales['Quantity'] * df_sales['Unit Price']
total_sales_amount = df_sales['Sales_Amount'].sum()

# Calculate average unit price
average_unit_price = df_sales['Unit Price'].mean()

# Calculate total quantity sold
total_quantity_sold = df_sales['Quantity'].sum()

# Calculate total gross profit
total_gross_profit = df_sales['Gross Profit'].sum()

# Print the summary
print("Brief Sales Data Summary:")
print(f"Total Sales Amount: ${total_sales_amount:,.2f}")
print(f"Average Unit Price: ${average_unit_price:,.2f}")
print(f"Total Quantity Sold: {total_quantity_sold}")
print(f"Total Gross Profit: ${total_gross_profit:,.2f}")
```
